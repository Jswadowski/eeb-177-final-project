## Read in canid range data.
### for each species get uniques and stick them into a dictionary with the earliest and latest occurrence.  


```python
data = open("/home/eeb177-student/Desktop/eeb-177/class-assignments/canids/cleaned_canids.csv", encoding = "ISO-8859-1")

infile = "cleaned_canids.csv"
with open(infile,'r') as ff:
    canid_recs = ff.readlines()[1:]

import os
from collections import defaultdict

#make a dictionary
species_ranges = defaultdict(list)
    

for line in canid_recs:     # each row is a list 
    species = line.split('","')[5]   #give me line element 5
    minage = line.split('","')[10]   #splitting on the literal "," to get rid of them
    maxage = line.split('","')[11]
    mean_age = (float(minage) + float(maxage))/2  
    #print(mean_age)
    
    species_ranges[species].append(mean_age)   #add species as keys and mean ages as elements to the list of ages
    #print(mean_age)
```

### get oldest and youngest ages of fossils for each species in species_ranges

```python
output = open("canid-ranges.csv", "w")    #making an empty file


for key in species_ranges.keys():
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    species = key
    #print(genus, key, minage, maxage)
    outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
    #print(outline)
    output.write(genus + "," + key + "," + 'minage' + "," + 'maxage' + "\n")
    output.write(outline)
    


#species_ranges
```

### sorting output so that species are listed alphabetically

### Write a function that does the following: takes a file name as argument, opens the file that is formatted like the cleaned version of the canid file we made from the shell in class, extracts the species name for each species with a latin binomial, ignores records that have only the genus recorded, populates a dictionary where the species are keys and the values are the ages of fossils known for each species, returns this dictionary.  

```python
import os
from collections import defaultdict


def using_filename(filename):
    with open(filename,'r') as ff:
        canid_recs = ff.readlines()[1:]
        species_ranges = defaultdict(list)
        for line in canid_recs:     # each row is a list 
            species = line.split('","')[5]   #give me line element 5
            
            is_two_words = 0 
            for char in species:
                
                if char == ' ':
                    is_two_words = 1
                if is_two_words > 0:
                    is_two_words = is_two_words + 1
                    
   
            #we want to add this entry
            # > 1 refers to: "Is there at least 1 character after the space to make it 2 words?"
            if is_two_words > 1:
                minage = line.split('","')[10]   #splitting on the literal "," to get rid of them
                maxage = line.split('","')[11]
                mean_age = (float(minage) + float(maxage))/2  
                #print(mean_age)
    
                #adding a key-value pair to a dictionary named species_ranges.
                #dict_name[key] = value
                species_ranges[species] = mean_age

    
    return species_ranges
    

hey = using_filename("cleaned_canids.csv")
for k,v in hey.items():
    print(k,v)
```

### Write a second function that takes a dictionary and output file name as arguments, opens a file for writing given the file name in the argument, determines the minimum and maximum fossil age for each species in the passed dictionary and writes to the outfile a csv with the following fields: genus, species, minage, maxage.  

```python
output = open("function2_ranges.csv", "w") 

def second(a_dict, output_file):   #dictionary and output file name are arguments
    output = open(output_file, "w")  #opens a file for writing
    for key in a_dict.keys():
        ages = a_dict[key]
        minage = min(ages)
        maxage = max(ages)
        
        genus = key.split(" ")[0]
        species = key
        outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
        output.write(outline)
    return a_dict

a_dict = using_filename("cleaned_canids.csv")
second(a_dict, "function2_ranges.csv")
with open("function2_ranges.csv", "r") as f:
    data2 = csv.reader(f, delimiter = ",")
    for rows in data2:
        print(rows)    #float is not iterable??
```
